```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50]

# Calculate variance
variance = np.var(data, ddof=0)  # ddof=0 for population variance; ddof=1 for sample variance

# Calculate standard deviation
std_dev = np.sqrt(variance)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
