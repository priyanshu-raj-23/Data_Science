```python
a = np.array(np.random.normal(5,1,1000))
b = np.array(np.random.normal(10,2,1000))
```

```python
plt.scatter(a,b)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(a, b)

def myfunc(a):
  return slope * a + intercept

mymodel = list(map(myfunc, a))
# sp = myfunc(20)
plt.scatter(a, b)
plt.plot(a, mymodel)
plt.show()
```
